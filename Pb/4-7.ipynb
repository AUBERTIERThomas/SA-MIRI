```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_evol(x, y, z):
    n_data = len(y)
    
    plt.figure(figsize=(12,6))
    for i in range(n_data):
        plt.loglog(x, y[i], 'o-', label=r"$size = {}$".format(z[i]))
    plt.xlabel("# Processes")
    plt.ylabel("Time (s)")
    plt.title("Evolution of execution time for different vector sizes")
    plt.grid(True, which="both")
    plt.legend()
    plt.show()
    
    plt.figure(figsize=(12,6))
    for i in range(n_data):
        plt.semilogx(x, [y[i][0]/yy for yy in y[i]], 'o-', label=r"$size = {}$".format(z[i]))
    plt.xlabel("# Processes")
    plt.ylabel("Speedup")
    plt.title("Evolution of speedup for different vector sizes")
    plt.grid(True, which="both")
    plt.legend()
    plt.show()
    
    plt.figure(figsize=(12,6))
    for i in range(n_data):
        plt.semilogx(x, [(y[i][0]*100)/(y[i][j]*x[j]) for j in range(len(y[i]))], 'o-', label=r"$size = {}$".format(z[i]))
    plt.xlabel("# Processes")
    plt.ylabel("Efficiency (%)")
    plt.title("Evolution of efficiency for different vector sizes")
    plt.grid(True, which="both")
    plt.legend()
    plt.show()
```

```python
processes = [1,2,4,8,16,32]
size_vect = [64,256,1024]

time = [[0.16, 4.29, 1.30, 0.12, 1.96, 0.48],
        [0.22, 3.14, 1.78, 1.14, 1.46, 2.34],
       [0.47, 3.55, 0.21, 0.22, 0.25, 0.45]]

plot_evol(processes, time, size_vect)
```
