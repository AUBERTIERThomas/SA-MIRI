```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_evol(N, kernel, memcpyHtoD, memcpyDtoH, malloc):
    
    plt.figure(figsize=(12,6))
    plt.loglog(N, [(memcpyDtoH[0]/(N[0]*N[0]))*c**(2) for c in N], '--', label=r"ORDER 2")
    plt.loglog(N, [(memcpyDtoH[0]/(N[0]*N[0]*N[0]*100))*c**(3) for c in N], '--', label=r"ORDER 3")
    plt.loglog(N, kernel, 'o-', label=r"Kernel MatMult")
    plt.loglog(N, memcpyHtoD, 'o-', label=r"Memcpy HtoD")
    plt.loglog(N, memcpyDtoH, 'o-', label=r"Memcpy DtoH")
    plt.loglog(N, malloc, 'o-', label=r"cudaMalloc")
    plt.xlabel("N")
    plt.ylabel("Time (ns)")
    plt.title("Execution size VS matrix size")
    plt.grid(True, which="both")
    plt.legend()
    plt.show()
```

```python
N = [32,64,128,256,512,1024,2048,4096,8192]
kernel = [4064, 6336, 10912, 20544, 71040, 553756, 4618655, 35607044, 336038035]
memcpyHtoD = [2592, 5504, 9536, 23744, 292638, 2260878, 6731664, 28081593, 151008836]
memcpyDtoH = [2848, 2880, 5024, 13024, 48223, 2360078, 9069632, 38040466, 145275822]
malloc = [90495550, 89427358, 99786795, 112170111, 100383888, 101104982, 88075134, 92258055, 94694548]

plot_evol(N, kernel, memcpyHtoD, memcpyDtoH, malloc)
```
